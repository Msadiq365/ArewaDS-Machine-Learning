```python

import matplotlib.pyplot as plt
import numpy as np

# Define the XRD pattern data
angles = np.linspace(0, 70, 1000)  # 2θ angles from 0 to 70 degrees
intensity_ss304 = np.random.normal(100, 10, size=1000)  # base intensity for ss304
intensity_joint = np.random.normal(200, 20, size=1000)  # base intensity for joint
intensity_zinc = np.random.normal(150, 15, size=1000)  # base intensity for Zinc

# Define peaks for the XRD patterns (arbitrary positions and intensities)
peaks_ss304 = [(15, 300), (30, 1000), (45, 500), (60, 700)]
peaks_joint = [(10, 800), (25, 1200), (35, 600), (50, 900)]
peaks_zinc = [(20, 700), (40, 1100), (55, 800)]

# Add peaks to the base intensity
for pos, height in peaks_ss304:
    intensity_ss304 += height * np.exp(-((angles - pos) ** 2) / 1)

for pos, height in peaks_joint:
    intensity_joint += height * np.exp(-((angles - pos) ** 2) / 1)

for pos, height in peaks_zinc:
    intensity_zinc += height * np.exp(-((angles - pos) ** 2) / 1)

# Create the plot
plt.figure(figsize=(12, 6))

# Plot XRD patterns
plt.plot(angles, intensity_ss304, color='blue', label='ss304 base metal')
plt.plot(angles, intensity_joint, color='red', label='Zn/ss304 joint')
plt.plot(angles, intensity_zinc, color='green', label='Zinc base metal')

# Annotate the IMCs
imc_annotations = {
    'ss304 base metal': [(15, 300, 'Fe3O4'), (30, 1000, 'Cr2O3'), (45, 500, 'Na2CrO4'), (60, 700, 'CaCl')],
    'Zn/ss304 joint': [(10, 800, 'CaCl'), (25, 1200, 'MgCl2'), (35, 600, 'ZnO'), (50, 900, 'Fe2O3')],
    'Zinc base metal': [(20, 700, 'ZnO'), (40, 1100, 'Zn(OH)2'), (55, 800, 'MgCl')]
}

for label, peaks in imc_annotations.items():
    for angle, height, imc in peaks:
        plt.text(angle, height + 50, imc, ha='center', va='bottom')

# Set labels and title
plt.xlabel('2θ, degrees')
plt.ylabel('Intensity, cps')
plt.title('XRD Analysis for the Corroded Samples of the Friction Welded Zn-SS')
plt.legend()

# Show the plot
#plt.grid(axis=False)
plt.show()
```
